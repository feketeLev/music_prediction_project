```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import glob

```

```python

# Define the path to your CSV files


csv_files = glob.glob('C:\\SpotifyAPI_tets\\Chart_download\\all\\*.csv')

print(csv_files)
# Dictionary to hold the aggregated streams
music_streams = {}
music_streams_by_url = {}
music_name_by_url = {}
# Process each CSV file
temp = 0
for file in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file)
    
    # Ensure the columns are named correctly
    df.columns = ['rank','url' ,'aritst_names' ,'track_name' ,'source' ,'peak_rank' ,'previous_rank' ,'weeks_on_chart', 'streams']
    
    # Aggregate the streams for each music track
    for index, row in df.iterrows():
        music = row['track_name']
        streams = row['streams']
        if music in music_streams:
            music_streams[music].append(streams)
        else:
            music_streams[music] = [streams]
    

# Convert the dictionary to a DataFrame for saving to a CSV
combined_df = pd.DataFrame(list(music_streams.items()), columns=['music', 'streams'])

# Save the combined data to a new CSV file
combined_df.to_csv('combined_music_streams.csv', index=False)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the combined CSV file
combined_df = pd.read_csv('combined_music_streams.csv')

# Extract the lengths of the stream lists
combined_df['num_data_points'] = combined_df['streams'].apply(lambda x: len(x.split(',')))

# Calculate the histogram data
hist_data, bin_edges = np.histogram(combined_df['num_data_points'], bins=range(1, combined_df['num_data_points'].max() + 2))

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(combined_df['num_data_points'], bins=range(1, combined_df['num_data_points'].max() + 2), edgecolor='black')
plt.title('Histogram of Number of Data Points per Song')
plt.xlabel('Number of Data Points')
plt.ylabel('Number of Songs')
plt.xticks(range(1, combined_df['num_data_points'].max() + 1))
plt.grid(axis='y')

# Save the plot as an image file
plt.savefig('histogram_num_data_points.png')

# Show the plot
plt.show()

# Print the histogram data
print("Histogram data (frequency counts):", hist_data)
print("Bin edges:", bin_edges)

# Optionally, save the histogram data to a CSV file
hist_df = pd.DataFrame({'Bin': bin_edges[:-1], 'Frequency': hist_data})
hist_df.to_csv('histogram_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file with average and standard deviation
df = pd.read_csv('sorted_music_streams_with_stats.csv')

# Ensure that the 'average' and 'std_dev' columns are correctly interpreted as floats
df['average'].astype(float)
df['std_dev'] = df['std_dev'].astype(float)

# Plot the averages with error bars for standard deviation
plt.figure(figsize=(12, 6))
plt.errorbar(df.index, df['average'], yerr=df['std_dev'], fmt='o', ecolor='r', capsize=5, linestyle='None', label='Average with Std Dev')

# Add labels and title
plt.xlabel('Music Index')
plt.ylabel('Average Stream Count')
plt.title('Average Stream Count with Standard Deviation for Each Music')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot as an image file
plt.savefig('average_with_std_dev_above15.png')

averages = df['average']
# Show the plot
plt.show()
```
